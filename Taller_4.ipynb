programa debe pedir al usuario ingresar una calificaci√≥n y su respectivo porcentaje o peso en cada
iteraci√≥n. Una vez ingresada una calificaci√≥n con su peso, el programa debe preguntar si desea
ingresar otra m√°s. Si el usuario responde "no", el programa calcula y muestra el promedio ponderado

Ejemplo:
Entrada:
* El programa pide ingresar la primera calificaci√≥n y su porcentaje:

o Porcentaje: 40%
* Luego, el programa pregunta si se desea agregar otra nota. El usuario responde "s√≠".
* El programa solicita la segunda calificaci√≥n:
o Nota: 4.2
o Porcentaje: 35%
* El programa pregunta de nuevo si se desea ingresar otra nota. El usuario responde "s√≠".
* El programa pide la √∫ltima calificaci√≥n:
o Nota: 4.0
o Porcentaje: 50%
* El programa pregunta si desea agregar m√°s notas. El usuario responde "no".
C√°lculo del promedio ponderado:

##Algoritmo

1. Inicio
2. Crear una lista vac√≠a notas para almacenar las notas y sus respectivos porcentajes.
3. Inicializar una variable add_more como verdadera para controlar el ciclo de ingreso de datos.
4. Mostrar el mensaje de bienvenida al usuario.
5. Mientras add_more sea verdadera:
Preguntar al usuario si desea agregar una nota.
Si la respuesta es "s":
6. Solicitar la nota.

9. Mostrar el mensaje de "Nota agregada con √©xito".
Si la respuesta es "n":
10. Cambiar add_more a falso.
11. Calcular el promedio ponderado:
12. Multiplicar cada nota por su respectivo porcentaje y sumar los productos.

```python
notas = []
add_more = True

print("Bienvenido al programa de c√°lculo del promedio ponderado")

while add_more:
    user_input = input("¬øDesea agregar una nota? (s/n): ").lower()

    if user_input == "s":
        nota = float(input("Ingrese la nota: "))
        porcentaje = float(input("Ingrese el porcentaje (en valor entero, ej. 40 para 40%): "))
        notas.append([nota, porcentaje])
        print("Nota agregada con √©xito")

    elif user_input == "n":
        add_more = False
        print("\nCalculando el promedio ponderado...\n")

        suma_productos = sum(nota * porcentaje for nota, porcentaje in notas)
        suma_porcentajes = sum(porcentaje for _, porcentaje in notas)

        if suma_porcentajes == 0:
            print("No se puede calcular el promedio con un porcentaje total de 0.")
        else:
            promedio_ponderado = suma_productos / suma_porcentajes
            print(f"El promedio ponderado es: {promedio_ponderado:.2f}")

    else:
        print("Valor no v√°lido, intenta nuevamente.")

    print("\n")
```
